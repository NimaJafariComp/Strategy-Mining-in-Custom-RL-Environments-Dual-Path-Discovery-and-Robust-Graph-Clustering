```python
import json
import pandas as pd

with open("final_filtered.json", "r") as f:
    episodes = json.load(f)

# Initialize outcome counters
count_treasure_win = 0   # e12
count_door_win = 0       # e9
count_not_won = 0        # e10

def label_event_sequence_clean(events):
    global count_treasure_win, count_door_win, count_not_won

    sequence = []
    key_collected = False
    explosive_collected = False

    for event in events:
        evt = event.get("event")

        if evt == "collect_item":
            item_type = event["item"][0]
            if item_type == "key":
                key_collected = True
                sequence.append("e1")
            elif item_type == "explosive":
                explosive_collected = True
                sequence.append("e2")

        elif evt == "failed_interaction":
            if event.get("item", [None])[0] == "key":
                sequence.append("e8")
            elif event.get("item", [None])[0] == "explosive":
                sequence.append("e7")

        elif evt == "interact":
            typ = event.get("type")
            what = event.get("item", [None])[0]

            if typ == "rock1":
                if what == "explosive":
                    sequence.append("e5")
                elif what == "key":
                    sequence.append("e8")

            elif typ == "rock2":
                if what == "explosive":
                    sequence.append("e5")
                elif what == "key":
                    sequence.append("e8")

            elif typ == "door":
                if what == "key":
                    sequence.append("e6")
                elif what == "explosive":
                    sequence.append("e7")

    if not key_collected:
        sequence.append("e3")
    if not explosive_collected:
        sequence.append("e4")

    # Determine outcome
    outcome_event = next((e for e in reversed(events) if e.get("event") == "game_won"), None)
    if outcome_event:
        if outcome_event.get("reason") == "treasure_collected":
            # Per your spec: whenever win is due to treasure, append both e11 and e12
            sequence.append("e11")  # player collects treasure (implicit via win reason)
            sequence.append("e12")  # player wins by collecting treasure
            count_treasure_win += 1
        else:
            sequence.append("e9")   # won by door (or any non-treasure reason)
            count_door_win += 1
    else:
        sequence.append("e10")      # game not won
        count_not_won += 1

    return sequence

# Process episodes
formatted_sequences = [label_event_sequence_clean(ep["events"]) for ep in episodes]
episode_ids = [f"ep_{i}" for i in range(len(episodes))]

# Save to CSV
df = pd.DataFrame({
    "episode_id": episode_ids,
    "sequence": formatted_sequences
})
df.to_csv("sequence_of_sets_formatted.csv", index=False)

# Print summary
print("Summary:")
print(f"  Total episodes               : {len(episodes)}")
print(f"  Won by treasure (e11+e12)    : {count_treasure_win}")
print(f"  Won by blue door (e9)        : {count_door_win}")
print(f"  Not won (e10)                : {count_not_won}")
print("Done! Saved to 'sequence_of_sets_formatted.csv'")
```
