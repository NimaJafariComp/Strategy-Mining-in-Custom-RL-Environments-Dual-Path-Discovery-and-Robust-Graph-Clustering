```python
import json
import pandas as pd

with open('final_run.json', 'r') as f:
    data = json.load(f)

# Remove 'move' events
for episode in data:
    if "events" in episode:
        episode["events"] = [e for e in episode["events"] if e.get("event") != "move"]

#Save
with open('final_run_clean.json', 'w') as f:
    json.dump(data, f, indent=2)

# Flatten all events into a DataFrame for preview
flattened_events = []
for ep_index, episode in enumerate(data):
    for event in episode.get("events", []):
        flat_event = event.copy()
        flat_event["episode_index"] = ep_index
        flattened_events.append(flat_event)

df = pd.DataFrame(flattened_events)


df.head(80)
```
